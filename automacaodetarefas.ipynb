# Passo a passo do Problema
- Passo1: buscar dados das ações automaticamente<br>
- Passo2: realizar as análises solicitadas pelo meu gestor<br>
- Passo3: enviar automaticamente o e-mail com o resultado da análise

```python
# importanto a biblioteca yfinance
import yfinance as yf

# inserindo o código da ação do Yahoo Finance
ticker = input('Digite o código da ação: ')

# puxando o histórico de 6 meses da ação
dados = yf.Ticker(ticker).history("6mo")

# puxa os dados da coluna que você definir após a variável dados
fechamento = dados.Close

# exibe os valores da ação em um gráfico de linha
fechamento.plot()
```

# Passo 2 - realizar análises da cotação
- Cotação máxima
- Cotação mínima
- Cotação atual

```python
fechamento
```

```python
# puxa o valor da última linha 
fechamento[-1]
```

```python
# puxa a cotação máxima e arredonda 
maxima = round(fechamento.max(), 2)

# puxa a cotação mínima e arredonda
minima = round(fechamento.min(), 2)

# puxa a cotação atual e arredonda
atual = round(fechamento[-1], 2)

# imprime os valores
print(maxima)
print(minima)
print(atual)
```

# Passo 3: enviar automaticamente o e-mail com o resultado da análise
- Abrir uma nova aba(ctrl + t)
- Digitar o endereço: www.gmail.com e dar um enter
- Clicar em escrever um novo e-mail
- Clicar no campo destinatários e insira o(s) endereço(s)
- Insira o assunto(tab)
- Insira o corpo do e-mail com as informações necessárias(tab)
- Clicar em enviar e-mail

```python
import pyautogui
import pyperclip
```

```python
# dar uma pausa de 5 segundos entre os passos
pyautogui.PAUSE = 5

# abrir uma nova aba (ctrl + t)
pyautogui.hotkey("ctrl", "t")

# digitar www.gmail.com e dar um enter
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

# clicar no botão Escreverabs
pyautogui.click(x=1036, y=208) # Puxa pelo comando pyautogui.click()

# digitar o destinatário
pyperclip.copy("tamirissoares.lopes@gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# digitar o assunto
pyperclip.copy("Análises diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# preencher o corpo do e-mail
mensagem = """Prezados Gestor,

Conforme solicitado, segue as análises solicitadas dos últimos seis meses da ação {ticker}:

Cotação máxima: R${maxima}
Cotação mínima: R${minima}
Cotação atual: R${atual}

Qualquer dúvida, fico à disposição.
At.te,

Phelippe Delalibera
Nome de Alguma empresa boa | Analista de Dados
"""
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

pyautogui.click(x=1451, y=994)
print("E-mail enviado com sucesso!")
```
